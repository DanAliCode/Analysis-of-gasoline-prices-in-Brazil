### 4. Você deve ter percebido que as colunas DATA INICIAL e DATA FINAL estão formatadas como string. Utilizando o método pd.to_datetime(), converta ambas colunas para Timestamp / Datetime.

```python
# Convertendo DATA INICIAL PARA DATETIME
df['DATA INICIAL'] = pd.to_datetime(df['DATA INICIAL'])
# Convertendo DATA FINAL PARA DATETIME
df['DATA FINAL'] = pd.to_datetime(df['DATA FINAL'])
```

### 5. Crie uma nova coluna para representar o ano e o mês(aaaa-mm), utilizando a coluna DATA FINAL como referência. 

```python
# Criando uma nova coluna com apenas o ano e o mes da Coluna DATA FINAL
df['ANO-MES'] = df['DATA FINAL'].apply(lambda x: '{}'.format(x.year)) + df['DATA FINAL'].apply(lambda x: '-{:02d}'.format(x.month))
df['ANO-MES']
```

### 7. Filtre o DataFrame para obter apenas dados da 'GASOLINA COMUM'. Grave em uma nova variável.

```python
# Criando a função para filtrar GASOLINA COMUM dentro da COLUNA PRODUTO
gc = df['PRODUTO'] == 'GASOLINA COMUM' # df database dentro de uma lista [df database[Produto = coluna] == 'GASOLINA COMUM' = item dentro da coluna]
dfgc = df[gc]
dfgc
```

### 8. Qual o preço médio de revenda da gasolina em agosto de 2008?

```python
dfgc[dfgc['ANO-MES'] == '2008-08']['PREÇO MÉDIO REVENDA'].mean()
```

### 9. Qual o preço médio de revenda da gasolina em maio de 2014 em São Paulo?

```python
dfgc[(dfgc['ANO-MES'] == '2014-05') & (dfgc['ESTADO'] == 'SAO PAULO')]['PREÇO MÉDIO REVENDA'].mean()
```

### 10. Você conseguiria descobrir em qual(quais) estado(s) a gasolina ultrapassou a barreira dos R$ 5,00? E quando isso ocorreu?

```python
# Filtrando quando o preço da gasolina ultrapassou R$ 5,00
dfgc[dfgc['PREÇO MÉDIO REVENDA'] > 5]
# Adicionando quais estados e quando ocorreu
dfgc[dfgc['PREÇO MÉDIO REVENDA'] > 5][['ESTADO', 'ANO-MES', 'PREÇO MÉDIO REVENDA']].head(10)
```

### 11. Qual a média de preço dos estados da região sul em 2012?

```python
df_aux = df[df['DATA FINAL'].apply(lambda x: x.year) == 2012]
df_aux[df_aux['REGIÃO'] == "SUL"]['PREÇO MÉDIO REVENDA'].mean()
```

### 12. Você conseguiria obter uma tabela contendo a variação percentual ano a ano para o estado do Rio de Janeiro?

```python
df['MES'] = df['DATA FINAL'].apply(lambda x: x.month)
df_rio = df[df['ESTADO'] == 'RIO DE JANEIRO']
df_month_rio = df_rio.groupby('ANO-MES')[['PREÇO MÉDIO REVENDA', 'MES']].last()
((df_month_rio[df_month_rio['MES'] == 12] / df_month_rio[df_month_rio['MES'] == 12].shift(1) - 1) * 100).rolling(3).sum()
```

### DESAFIO: Crie uma tabela contendo uma serie temporal com a diferença absoluta e percentual entre os valores mais baratos e caros. Apresente também ao lado os estados na qual os maiores e menores preços foram registrados.

```python

df_max = df.groupby("ANO-MES")["PREÇO MÉDIO REVENDA"].max()
df_min = df.groupby("ANO-MES")["PREÇO MÉDIO REVENDA"].min()

```

```python
idx_max = df.groupby("ANO-MES")["PREÇO MÉDIO REVENDA"].idxmax()
idx_min = df.groupby("ANO-MES")["PREÇO MÉDIO REVENDA"].idxmin()
```

```python
df_diff = pd.DataFrame()
```

```python
df_diff['abs_diff'] = df_max - df_min
df_diff['percent_diff'] = (df_max - df_min) / df_min * 100
df_diff['max'] = df_max
df_diff['min'] = df_min
df_diff
```

```python
df.loc[idx_max, :][["ESTADO", "PREÇO MÉDIO REVENDA", "ANO-MES"]]
```

```python

df_diff["ESTADO_MAX"] = df.loc[idx_max, :]['ESTADO'].values
df_diff["ESTADO_MIN"] = df.loc[idx_min, :]['ESTADO'].values
```
